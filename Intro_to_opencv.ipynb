```python
fix_img_1 = cv2.resize(fix_img,(600,500)) # 600 -width & 500- height
```

```python
plt.imshow(fix_img_1) # display resized image
plt.show() # show the plotted image
```

```python
fix_img_1.shape # get shape of fixed image_1
```

```python
w_ratio = 0.3 # width ratio
h_ratio = 0.3 # height ratio
```

```python
fix_img_2 = cv2.resize(fix_img,(0,0), fx=w_ratio, fy=h_ratio) # resize using ratio
```

```python
fix_img_2.shape # 50% of the fix_img shape 
```

```python
plt.imshow(fix_img_2) # display resized image using ratio
plt.show() # show the plotted image
```

```python
plt.imshow(fix_img) # display the fixed image
plt.show() # Show the plotted image of original fixed image
```

```python
img3 = cv2.flip(fix_img_2,0) # flip the image vertically
plt.imshow(img3) # flip the image vertically
plt.show() # display the flipped image
```

```python
cv2.imwrite('flipped_image.jpg',img3) # save the image to local system
```

```python
import os # to get the absolute path
cv2.imwrite('flipped_image.jpg', img3) # save the image to local system
print("Saved at:", os.path.abspath('flipped_image.jpg')) # print the absolute path where image is saved
```
