```python
image = cv2.imread(r"C:\Users\Ramya\OneDrive\Desktop\car image.jpg")
resized_image = imutils.resize(image , width=6000)
# Convert BGR (OpenCV default) to RGB for displaying
image_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
 #Display using matplotlib
plt.figure(figsize=(8,6))
plt.imshow(image_rgb)
plt.axis('off')  # hides the axes
plt.title("Original Image")
plt.show()
```

```python
# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display in Jupyter
plt.figure(figsize=(8,6))
plt.imshow(gray_image, cmap='gray')  # cmap='gray' for grayscale
plt.axis('off')
plt.title("Grayscale Image")
plt.show()
```

```python
# Apply bilateral filter
gray_image_smooth = cv2.bilateralFilter(gray_image, 11, 17, 17)

# Display in Jupyter
plt.figure(figsize=(8,6))
plt.imshow(gray_image_smooth, cmap='gray')  # use 'gray' for grayscale
plt.axis('off')
plt.title("Smoothened Image")
plt.show()
```

```python
# Apply Canny edge detection
edged = cv2.Canny(gray_image, 30, 200)

# Display in Jupyter
plt.figure(figsize=(8,6))
plt.imshow(edged, cmap='gray')  # Canny edges are grayscale
plt.axis('off')
plt.title("Edged Image")
plt.show()
```

```python
# Find contours
cnts, _ = cv2.findContours(edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Copy original image to draw contours
image1 = image.copy()
cv2.drawContours(image1, cnts, -1, (0, 255, 0), 3)

# Convert BGR to RGB for matplotlib display
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

# Display contours in Jupyter
plt.figure(figsize=(10,8))
plt.imshow(image1_rgb)
plt.axis('off')
plt.title("Contours")
plt.show()
```

```python
# Sort contours by area and take top 30
cnts = sorted(cnts, key=cv2.contourArea, reverse=True)[:30]

# Copy original image to draw contours
image2 = image.copy()
cv2.drawContours(image2, cnts, -1, (0, 255, 0), 3)

# Convert BGR to RGB for matplotlib
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# Display top 30 contours
plt.figure(figsize=(10,8))
plt.imshow(image2_rgb)
plt.axis('off')
plt.title("Top 30 Contours")
plt.show()
```

```python
i = 7
screenCnt = None

for c in cnts:
    perimeter = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.018 * perimeter, True)
    
    # Check for quadrilateral
    if len(approx) == 4:
        screenCnt = approx
        
        # Get bounding box and crop the detected region
        x, y, w, h = cv2.boundingRect(c)
        new_img = image[y:y+h, x:x+w]
        
        # Save the cropped image
        cv2.imwrite(f'./{i}.png', new_img)
        i += 1
        
        # Display the cropped region inline in Jupyter
        new_img_rgb = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(6,4))
        plt.imshow(new_img_rgb)
        plt.axis('off')
        plt.title(f"Cropped Region {i-1}")
        plt.show()
        
        break  # exit after finding the first quadrilateral
```

```python
# Draw the detected license plate contour on the original image
cv2.drawContours(image, [screenCnt], -1, (0, 255, 0), 3)

# Convert BGR → RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display in Jupyter
plt.figure(figsize=(10,6))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Image with Detected License Plate")
plt.show()
```
